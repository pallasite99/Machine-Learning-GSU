```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Given dataset
X = np.array([
    [5.0, 4.8],
    [1.0, 1.4],
    [4.4, 5.8],
    [3.8, 4.4],
    [6.2, 6.0],
    [4.6, 5.4],
    [4.0, 3.2],
    [2.0, 2.2],
    [3.0, 3.2],
    [2.2, 1.8]
])
```

```python
# Scatter plot of the original dataset
plt.scatter(X[:, 0], X[:, 1], label='Original Data')
plt.xlabel('x_i')
plt.ylabel('y_i')
plt.title('Scatter Plot of Original Data')
plt.legend()
plt.show()
```

```python
# Step 2: Normalize the data
mean_X = np.mean(X, axis=0)
X_centered = X - mean_X
```

```python
# Scatter plot of the normalized data
plt.scatter(X_centered[:, 0], X_centered[:, 1], label='Normalized Data', color='red')
plt.xlabel('x_i')
plt.ylabel('y_i')
plt.title('Scatter Plot of Normalized Data')
plt.legend()
plt.show()
```

```python
# Step 3: Compute the covariance matrix
cov_matrix = np.cov(X_centered.T)
print("Covariance Matrix:\n", cov_matrix)
```

```python
# Step 4: Compute eigenvalues and eigenvectors
eig_values, eig_vectors = np.linalg.eig(cov_matrix)
print("Eigenvalues:\n", eig_values)
print("Eigenvectors:\n", eig_vectors)
```

```python
# Step 5: Select the principal component (largest eigenvalue)
principal_component = eig_vectors[:, np.argmax(eig_values)]
print("Principal Component:\n", principal_component)
```

```python
# Step 6: Project data onto the principal component
X_pca = X_centered @ principal_component.reshape(-1, 1)
X_reconstructed = X_pca @ principal_component.reshape(1, -1) + mean_X
```

```python
# Step 7: Plot reconstructed data vs original data
plt.scatter(X[:, 0], X[:, 1], label='Original Data', alpha=0.5)
plt.scatter(X_reconstructed[:, 0], X_reconstructed[:, 1], label='Reconstructed Data', color='green')
plt.xlabel('x_i')
plt.ylabel('y_i')
plt.title('Comparison of Original and Reconstructed Data')
plt.legend()
plt.show()
```
